```python


class ControlCharts:
    """
    Sistema de Cartas de Controle para Dataset Real de Manufatura
    """

    def __init__(self, data):
        self.data = data.copy()
        self.subgroup_size = None
        print("üìä Sistema inicializado!")

    def set_subgroup_size(self, n):
        self.subgroup_size = n

    def create_subgroups(self, data_column):
        if self.subgroup_size is None:
            raise ValueError("‚ùå Defina o tamanho do subgrupo primeiro!")

        data = self.data[data_column].dropna()
        n_groups = len(data) // self.subgroup_size
        subgroups = []

        for i in range(n_groups):
            start_idx = i * self.subgroup_size
            end_idx = start_idx + self.subgroup_size
            subgroup_data = data.iloc[start_idx:end_idx]
            subgroups.append({
                'subgroup': i + 1,
                'mean': subgroup_data.mean(),
                'range': subgroup_data.max() - subgroup_data.min(),
                'std': subgroup_data.std()
            })

        return pd.DataFrame(subgroups)

    def xbar_r_chart(self, data_column, subgroup_size=5):
        print(f"üéØ ANALISANDO VARI√ÅVEL: {data_column}")
        print("=" * 50)

        try:
            self.set_subgroup_size(subgroup_size)
            subgroups_df = self.create_subgroups(data_column)

            # C√°lculos estat√≠sticos
            xbar_mean = subgroups_df['mean'].mean()
            r_mean = subgroups_df['range'].mean()

            # Constantes para cartas de controle
            constants = {5: {'A2': 0.577, 'D3': 0, 'D4': 2.114}}
            A2 = constants[subgroup_size]['A2']
            D3 = constants[subgroup_size]['D3']
            D4 = constants[subgroup_size]['D4']

            # Limites de controle
            ucl_xbar = xbar_mean + A2 * r_mean
            lcl_xbar = xbar_mean - A2 * r_mean
            ucl_r = D4 * r_mean
            lcl_r = D3 * r_mean

            # Criar gr√°ficos
            import matplotlib.pyplot as plt
            fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

            # Carta X-barra
            ax1.plot(subgroups_df['subgroup'], subgroups_df['mean'], 'bo-', linewidth=1, markersize=4)
            ax1.axhline(xbar_mean, color='red', linestyle='-', linewidth=2, label='Linha Central')
            ax1.axhline(ucl_xbar, color='red', linestyle='--', linewidth=1.5, label='LSC')
            ax1.axhline(lcl_xbar, color='red', linestyle='--', linewidth=1.5, label='LIC')
            ax1.set_title(f'CARTA X-BARRA - {data_column} (Dataset Real)')
            ax1.set_ylabel('M√©dia')
            ax1.legend()
            ax1.grid(True, alpha=0.3)

            # Carta R
            ax2.plot(subgroups_df['subgroup'], subgroups_df['range'], 'go-', linewidth=1, markersize=4)
            ax2.axhline(r_mean, color='red', linestyle='-', linewidth=2, label='Linha Central')
            ax2.axhline(ucl_r, color='red', linestyle='--', linewidth=1.5, label='LSC')
            ax2.axhline(lcl_r, color='red', linestyle='--', linewidth=1.5, label='LIC')
            ax2.set_title(f'CARTA R - {data_column} (Dataset Real)')
            ax2.set_ylabel('Amplitude')
            ax2.set_xlabel('Subgrupo')
            ax2.legend()
            ax2.grid(True, alpha=0.3)

            plt.tight_layout()
            plt.show()

            # Estat√≠sticas
            print(f"üìä RESULTADOS PARA {data_column}:")
            print(f"   ‚Ä¢ Linha Central X-barra: {xbar_mean:.4f}")
            print(f"   ‚Ä¢ Linha Central R: {r_mean:.4f}")
            print(f"   ‚Ä¢ LSC: {ucl_xbar:.4f}")
            print(f"   ‚Ä¢ LIC: {lcl_xbar:.4f}")
            print(f"   ‚Ä¢ Subgrupos analisados: {len(subgroups_df)}")

            return subgroups_df

        except Exception as e:
            print(f"‚ùå Erro na an√°lise: {e}")
            return None

print("‚úÖ SISTEMA DE CARTAS DE CONTROLE PRONTO!")
```

```python
print("üéØ PRIMEIRA AN√ÅLISE")
print("=" * 45)

# Inicializar
control_system = ControlCharts(real_data)

# Escolher a primeira vari√°vel num√©rica para an√°lise
if numeric_columns:
    first_variable = numeric_columns[0]
    print(f"üìè ANALISANDO: {first_variable}")
    print(f"üìä Descri√ß√£o: {real_data[first_variable].describe()}")

    # Gerar carta de controle
    results = control_system.xbar_r_chart(first_variable, subgroup_size=5)

    print("\n‚úÖ AN√ÅLISE CONCLU√çDA!")
else:
    print("‚ùå Nenhuma vari√°vel num√©rica encontrada para an√°lise")
```
